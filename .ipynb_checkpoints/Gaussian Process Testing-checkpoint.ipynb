```python
import numpy as np
from matplotlib import pyplot as plt
import scipy
```

## Create Kernel Function

```python
σ2 = 1
L = 1

domain = np.linspace(0,2*np.pi,50)

def kernel(X1,X2):
    k = lambda x1, x2 : σ2 * np.exp(-(x1 - x2)**2 / (2*L**2))
    A,B = np.meshgrid(X1,X2)
    return k(A,B)

K = kernel(domain,domain)

plt.figure(figsize=(3,2))
plt.imshow(K, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Value')
plt.show()
```

```python
vals = [0,.3]

mu = np.array([0,0])
cov = kernel(vals,vals)

plt.figure(figsize=(3,3))
sample = np.random.multivariate_normal(mean=mu,cov=cov,size=1000)
plt.scatter(sample[:,0],sample[:,1],s=5)
plt.title(f'Distributions: {vals}')
plt.show()
```

## Sample Functions (with no data)

```python
num_samples = 5

functions = np.random.multivariate_normal(mean=np.zeros_like(domain),cov=K,size=num_samples)

plt.figure(figsize=(5,3))
for i in range(num_samples):
    plt.plot(domain,functions[i,:],marker='o',linestyle='-',markersize=3)
plt.show()
```

```python
def kernel(X1,X2):
    σ2 = 1
    L = 1
    k = lambda x1, x2 : np.array([[σ2 * np.exp(-(v1 - v2)**2 / (2*L**2)) for v2 in x2] for v1 in x1])
    return k(X1,X2)
```

```python
data_size = 6
posterior_samples = 50
function_samples = 5
domain = [0,2*np.pi]

# Create Original Dataset
X1 = np.random.uniform(domain[0]+1,domain[1]-1,data_size-2)
f = lambda x : np.sin(x)
Y1 = f(X1)

# Sample From Posterior
X2 = np.linspace(domain[0],domain[1],posterior_samples)

K_11 = kernel(X1,X1)
K_12 = kernel(X1,X2)
K_22 = kernel(X2,X2)

solved = scipy.linalg.solve(K_11, K_12, assume_a='pos').T

# Solve for New Multivariate Gaussian Distribution Parameters
μ2 = solved @ Y1
K2 = K_22 - solved @ K_12

var2 = np.diag(K2)

Y2 = np.random.multivariate_normal(mean=μ2,cov=K2,size=function_samples)
```

```python
plt.figure(figsize=(8,3))

plt.plot(X2,f(X2),'b--',label='$sin(x)$')
plt.scatter(X1,Y1,s=15,c='k',label='$(x,y)$')
plt.fill_between(X2,μ2-2*var2,μ2+2*var2,color='red',alpha=0.25,label='95% Confidence')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gaussian Process')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,3))
for i in range(function_samples):
    plt.plot(X2,Y2[i,:],marker='o',linestyle='-',markersize=3)
plt.show()
```
